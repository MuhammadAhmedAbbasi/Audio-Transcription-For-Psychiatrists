```python
import librosa
```

```python
import whisperx
import gc
```

```python
device = "cpu"
batch_size = 4 # reduce if low on GPU mem
compute_type = "int8" # change to "int8" if low on GPU mem (may reduce accuracy)
```

```python
audio_file = r"C:\Users\Administrator\Desktop\Backend-Algorithm-LLM\Audio_transcription_files\tmpw19f5zrc.wav"
```

```python
full_dir_path = r'C:\Users\Administrator\Desktop\Backend-Algorithm-LLM\Algorithm\audio_text_llm\SenseVoiceSmall\audio_sample'
audio_array, sampling_rate = librosa.load(audio_file)
audio_array = librosa.resample(audio_array,orig_sr= sampling_rate, target_sr = 16000)
```

```python
model = whisperx.load_model("large-v2", device,compute_type=compute_type, language= 'zh')
```

```python
result = model.transcribe(audio_array, batch_size=batch_size)
```

```python
print(result["segments"]) # before alignment
```

```python
model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)
```

```python
result = whisperx.align(result["segments"], model_a, metadata, audio_array, device, return_char_alignments=False)
```

```python
print(result["segments"]) # after alignment
```

```python
diarize_model = whisperx.DiarizationPipeline(use_auth_token='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', device=device)
```

```python
diarize_segments = diarize_model(audio_array,min_speakers=2, max_speakers=2)
```

```python
result = whisperx.assign_word_speakers(diarize_segments, result)
```

```python
print(diarize_segments)
```

```python
print(result["segments"]) # segments are now assigned speaker IDs
```

```python
speaker_0 = []
speaker_1 = []
final_transcription = []

for num, segment in enumerate(result["segments"]):
    sub_0 = []
    sub_1 = []
    for word_info in segment.get('words', []):
        try:
            if word_info['speaker'] == 'SPEAKER_01':
                sub_1.append(word_info['word'])
            else:
                sub_0.append(word_info['word'])
        except KeyError:
            print("No speaker info available")

    sub_0_sen = " ".join(sub_0)
    sub_1_sen = " ".join(sub_1)
    
    # Only append if the sentence is not empty
    if sub_0_sen:
        speaker_0.append(sub_0_sen)
        final_transcription.append(f"Speaker 0: {sub_0_sen}")
    if sub_1_sen:
        speaker_1.append(sub_1_sen)
        final_transcription.append(f"Speaker 1: {sub_1_sen}")
```
